# Números complejos

## Introducción y propiedades

**Definición 1.1.** Un número complejo es un par ordenado de números reales $(a,b)$.

Dado un número complejo $z = (a, b)$: 
- $a = Re(z)$
- $b = Im(z)$

**Definición 1.2** Sean $(a, b)$ y $(c, d)$ complejos:

- $(a, b) = (c, d) \iff a = b \land c = d$
- $(a, b) + (c, d) = (a + b, c + d)$
- $(a, b) \cdot (c, d) = (ac - bd, ad + bc)$

**Neutros del cuerpo $\mathbb{C}$**

* Neutro de suma: $(0, 0)$
* Neutro de producto: $(1, 0)$

**Ejercicio 1.1**

Verificar que, dado un complejo $(a, b)$:
1. Su opuesto está dado por $(-a, -b)$
2. Si $(a, b) \neq (0, 0)$ su inverso es $(\frac{a}{a^2+b^2}, \frac{-b}{a^2+b^2})$

*1)* Sea $(c, d)$ un número complejo tal que $(a, b) + (c, d) = (0, 0)$. Entonces

*(Def 1.2)* $ \implies a + c = 0$ y $b + d = 0$ $\implies a = -c$ y $b = -d$

Por lo tanto $(c, d) = (-a, -b)$ y es el opuesto de $(a, b)$. $\square$

```python
(1 + 2j) + (-1 - 2j) == (0 + 0j) == 0j
```

*2)* Sea $(c, d)$ un número complejo tal que $(a, b) * (c, d) = (1, 0)$. Entonces

*(Def 1.2)* $ \implies ac - bd = 1$ y $ad + bc = 0$

De la segunda ecuación $\implies d = \frac{-bc}{a}$

Reemplazo en la primer ecuación:

$$
\begin{align*}
ac + \frac{b^2c}{a} = c \cdot (a + \frac{b^2}{a}) = c \cdot (\frac{a^2 + b^2}{a}) &= 1
\\
c &= \frac{a}{a^2 + b^2}
\end{align*}
$$

Reemplazo a $c$ en la ecuación que define a $d$ y obtengo:

$$
d = \frac{-b}{a} \frac{a}{a^2 + b^2} = \frac{-b}{a^2 + b^2}
$$

Por lo tanto $(c, d)$ tiene la forma:

$$
(c, d) = (\frac{a}{a^2 + b^2}, \frac{-b}{a^2 + b^2})
$$

$\square$

```python
def inverse(z):
    modulo = z.real ** 2 + z.imag ** 2
    return (z.real - z.imag * 1j) / modulo

z = (2 + 4j)
z * inverse(z) == 1
```

**Definición número $i$** Sea el número complejo $(0, 1)$ notamos que si lo multiplicamos por símismo obtenemos:

$$(0, 1) \cdot (0, 1) = (0\cdot0 - 1\cdot1, 0\cdot1 + 1\cdot0) = -1$$

A este número lo llamamos $i$. Podemos por lo tanto pasar a notar un número $(a, b)$ como $z = a + bi$.

**Definición 1.3** Sea $z = a + bi$, Llamamos *módulo* de $z$ a su distancia al origen $\|z\| = \sqrt{a^2 + b^2}$, y llamamos *argumento* al ángulo $\phi$ que se forma con el eje horizontal. El *argumento* se puede calcular como $\phi = \arctan{\frac{b}{a}}$.

Hay ciertas excepciones si $(a > 0 \land b < 0) \lor (a < 0 \land b < 0)$, en cuyo caso se tiene que ajustar el *argumento* sumando o restando $\pi$

**Propocisión 1.1** Sean $z$ y $w$ números complejos, entonces:

1. $\|z\| \geq 0 \iff z \neq 0$
2. $\|z\| = 0 \iff z = 0$
3. $\|Re(z)\| \leq \|z\|$
4. $\|Im(z)\| \leq \|z\|$
5. $\|zw\| = \|z\| \cdot \|w\|$
5. $\|\frac{z}{w}\| = \frac{\|z\|} {\|w\|}$

Además tenemos que si $z = (a, b)$ y su argumento es $\phi$:

1. $a = Re(z) = \|z\| \cdot \cos \phi$
1. $b = Im(z) = \|z\| \cdot \sin \phi$

**Definición 1.4** Sea $z = (a, b)$ complejo, su *conjugado* es $\overline{z} = a - bi$

**Propocisión 1.2**: Propiedades de números complejos

Sean $z$ y $w$ dos números complejos:

1. $\overline{z + w} = \overline{z} + \overline{w}$
2. $\overline{zw} = \overline{z} \cdot \overline{w}$
3. $z + \overline{z} = 2 Re(z)$
4. $z - \overline{z} = 2i \cdot Im(z)$
5. $\bar{\bar{z}} = z$
6. $z \cdot \overline{z} = \|z\|^2$

**Notación polar:** Sean $z$ un número complejo y $\theta$ entre $0$ y $2\pi$:

$$
z = \|z\| \cdot (\cos{\theta} + i\sin{\theta})
$$

## Exponencial compleja

**Definición 1.5** Dado un complejo $z = a + bi$, se define exponencial compleja como:

$$
e^z = e^a \cdot (\cos{b} + i\sin{b})
$$

**Notación polar compacta** Sabiendo que un imaginario puro tiene la forma $e^{ib} = \cos{b} + i\sin{b}$, sea $z$ un complejo de módulo $\rho$ y ángulo $\theta$:

$$
z = \rho \cdot e^{i\theta}
$$

*Ejemplos:*

1. $1 = e^{i0}$

2. $i = e^{i \frac{\pi}{2}}$

3. $-i = e^{i \frac{3\pi}{2}} = e^{i \frac{-\pi}{2}}$

4. $1 + i = \sqrt{2} \cdot e^{i \frac{\pi}{4}}$

*Representación geométrica:*

Las operaciones de suma entre complejos son muy intuitivas, pero las de producto no tienen una explicación inmediata. Representando los números complejos en esta notación facilita ver la **representación geométrica del producto** de dos números complejos.

Sean dos complejos $z1 = \rho_1 \cdot e^{i\theta_1}$ y $z2 = \rho_2 \cdot e^{i\theta_2}$, tenemos que:

$$z1 \cdot z2 = \rho_1 \rho_2 \cdot e^{i (\theta_1 + \theta_2)}$$

Es decir:

1. Se multiplican los módulos
2. Se suman los argumentos

```python
z1 = np.sqrt(2) * np.e ** (np.pi / 4 * 1j)
z2 = 2 * np.e ** (np.pi / 2 * 1j)
values = np.array([z1, z2, z1 * z2])

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(values.real, values.imag)
[plt.plot([0, v.real], [0, v.imag]) for v in values]
plt.show()
```

## Raíces complejas

Dado un complejo $w_0$, queremos calcular todos los $z \in \mathbb{C}$ tal que $z^n = w_0$. Si escribimos a ambos en notación polar tenemos:

$$
\rho^n \cdot e^{in\theta} = \rho_0 \cdot e^{i\theta_0}
$$


*Ejemplo*

Si queremos calcular la raíz n-ésima de 1, tenemos que $z^n = 1$ y por lo tanto $\rho^n \cdot e^{in\theta} = 1$. Entonces:

1. $\rho^n = 1 \implies \rho = 1$ pues $\rho \in \mathbb{N}$
2. $n\theta = 2k\pi$ con $k \in \mathbb{Z}$, y obtendremos $n$ soluciones de la forma

$$\theta_k = \frac{2k\pi}{n}$$

**Generalización**

Si queremos calcular la raíz n-ésima de $w_0$, obtendremos:

$$
\begin{align*}
\rho^n &= \rho_0 \implies \rho = \sqrt[n]{\rho_0}
\\
n\theta &= \theta_0 + 2k\pi \implies \theta = \frac{\theta_0}{n} + \frac{2k\pi}{n}\ ,\ k = 0, ..., n - 1
\end{align*}
$$

**Representación geométrica**

Representemos $z^7 = 1$. Entonces:

1. $\rho = 1$
2. $\theta = \frac{0}{7} + \frac{2k\pi}{7}\ ,\ k = 0, ..., 6$

```python
import cmath

def plot_complex_roots(w0=1 + 0j, n=2):
    rho_0, theta_0 = cmath.polar(w0)
    rho = np.power(rho_0, 1 / n)

    values = np.array([rho * np.e ** (1j * theta_0 / n + 1j * 2 * k * np.pi / n) for k in range(n)])
    fig, ax = plt.subplots(figsize=(5,5))
    ax.scatter(values.real, values.imag)
    [plt.plot([0, v.real], [0, v.imag]) for v in values]
    ax.add_artist(plt.Circle((0, 0), rho, alpha=0.3))

plot_complex_roots(w0=1, n=7)
```

Representemos $z^{10} = 1$. Entonces:

1. $\rho = 1$
2. $\theta = \frac{0}{10} + \frac{2k\pi}{10}\ ,\ k = 0, ..., 9$

```python
plot_complex_roots(w0=1, n=10)
```

Representemos $z^{6} = 8 \cdot (\sqrt(3) -i)$. Entonces:

0. $w_0 = 8\sqrt{3} - 8i$
1. $\rho_0 = \sqrt{(8\sqrt{3})^2 + 8^2} = \sqrt{64 * 3 + 64} = \sqrt{256} = 16$
2. $\theta_0 = arctan(\frac{-8}{8\sqrt{3}}) = arctan(\frac{-1}{\sqrt{3}})$

$$
\begin{align*}
\rho &= \sqrt[6]{\rho_0}
\\
\theta &= \frac{\theta_0}{6} + \frac{2k\pi}{6}\ ,\ k = 0, ..., 5
\end{align*}
$$

```python
plot_complex_roots(w0=8 * (np.sqrt(3) - 1j), n=6)
```
